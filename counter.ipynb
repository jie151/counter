```python
import firebase_admin
from firebase_admin import credentials
from firebase_admin import firestore

if not firebase_admin._apps:
    cred = credentials.Certificate("./service-account.json")
    firebase_admin.initialize_app(cred)
    
db = firestore.client()
```

```python
docs = db.collection(u'cguscholar').stream()
count = 0
for doc in docs:
    print(doc.id)
    count=count+1
count
```

```python
quantitycalculationlist = []
docs = db.collection(u'Label-Domain').where(u'updateTime', u'>', '').stream()

for doc in docs:
    label_calculation = {}
    
    #if label['userID'] is not empty
    if doc.exists:
        docTemp = doc.to_dict()
        labelIDcount = docTemp['userID']
        count = 0
        
        #label calculation
        for _labelIDcount in labelIDcount:
            count = count +1
        label_calculation['label name'] = doc.id
        label_calculation['label calculation'] = count
    quantitycalculationlist.append(label_calculation)
quantitycalculationlist
```

```python
label_info = ['label name', 'label calculation']
```

```python
import datetime

def currentTime():
    now = datetime.datetime.now()
    currentTime = now.strftime("%Y-%m-%d")
    return str(currentTime)
```

```python
import csv
with open('./labelcountfile/ '+currentTime()+'.csv', 'w') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames = label_info)
    writer.writeheader()
    writer.writerows(quantitycalculationlist)
```
